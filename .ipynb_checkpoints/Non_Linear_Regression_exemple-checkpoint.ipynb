### Introduction 


Premièrement, la régression non linéaire est une méthode permettant de modéliser une relation non linéaire entre les
variable dépendante et un ensemble de variables indépendantes.
Deuxièmement, pour qu'un modèle soit considéré comme non linéaire, y ̂à prédire doit être une fonction non linéaire de la
paramètres θ, pas nécessairement les caractéristiques x.
Lorsqu'il s'agit d'une équation non linéaire, elle peut avoir la forme d'une exponentielle, logarithmique,
et logistique, ou de nombreux autres types.

La régression non linéaire, un modèle est non linéaire par paramètres.
Contrairement à la régression linéaire, nous ne pouvons pas utiliser la méthode ordinaire des "moindres carrés" pour ajuster
les données en régression non linéaire, et en général, l'estimation des paramètres n'est pas facile.

Permettez-moi de répondre ici à deux questions importantes : Premièrement, "Comment puis-je savoir si un problème est linéaire
ou non linéaire de manière simple" ?
Pour répondre à cette question, nous devons faire deux choses :
La première consiste à déterminer visuellement si la relation est linéaire ou non linéaire.
Il est préférable de tracer des tracés bivariés des variables de sortie avec chaque variable d'entrée.
De plus, vous pouvez calculer le coefficient de corrélation entre les variables indépendantes et les variables dépendantes,
et si, pour toutes les variables, il est égal ou supérieur à 0,7, il y a une tendance linéaire, et donc
n'est pas approprié pour s'adapter à une régression non linéaire.
La deuxième chose à faire est d'utiliser une régression non linéaire au lieu d'une régression linéaire lorsque
nous ne pouvons pas modéliser avec précision la relation avec les paramètres linéaires.
La deuxième question importante est la suivante : "Comment dois-je modéliser mes données, si elles affichent des paramètres non linéaires sur un diagramme de dispersion ?"
Pour résoudre ce problème, il faut utiliser soit une régression polynomiale, soit une régression non linéaire
modèle de régression, ou "transformer" vos données.


Si les données montrent une tendance curviligne, alors la régression linéaire ne donnera pas de résultats très précis par rapport à une régression non linéaire car, comme son nom l'indique, la régression linéaire suppose que les données sont linéaires. 

### Importation des bibliothèques

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = 2*(x) + 3
y_noise = 2 * np.random.normal(size=x.size) # l'ajout d'un bruit gaussiene à notre variable à prédire 
y_data=y+y_noise
plt.figure(figsize=(15,6))
plt.plot(x, y_data,'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

Les régressions non linéaires sont une relation entre des variables indépendantes $x$ et une variable dépendante $y$ qui donnent des données modélisées par une fonction non linéaire. Essentiellement, toute relation qui n'est pas linéaire peut être qualifiée de non linéaire, et est généralement représentée par le polynôme de $k$ degrés (puissance maximale de $x$). 

$$ \ y = a x^3 + b x^2 + c x + d \ $$

Les fonctions non linéaires peuvent comporter des éléments comme les exponentielles, les logarithmes, les fractions et autres. Par exemple : $$ y = \log(x)$$

Une fonction cubique : 

```python
x = np.arange(-5.0, 5.0, 0.1)

y = 1*(x**3) + 1*(x**2) + 1*x + 3
y_noise = 20 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

Comme vous pouvez le voir, cette fonction a $x^3$ et $x^2$ comme variables indépendantes. De plus, le graphique de cette fonction n'est pas une ligne droite sur le plan 2D. Il s'agit donc d'une fonction non linéaire.

Une fonction quadratique : 

$$ Y = X^2 $$

```python
x = np.arange(-5.0, 5.0, 0.1)

y = np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

La fonction exponentielle : 

Une fonction exponentielle avec la base c est définie par $$ Y = a + b c^X$$ où b ≠0, c > 0 , c ≠1, et x est un nombre réel quelconque. La base, c, est constante et l'exposant, x, est une variable. 


```python
X = np.arange(-5.0, 5.0, 0.1)

Y= np.exp(X)

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

Une fonction logarithmique

La réponse $y$ est le résultat de l'application d'une fonction logarithmique de la variable d'entrée $x$ à la variable de sortie $y$. C'est l'une des formes les plus simples de __log()__ : c'est-à-dire $$ y = \log(x)$$

Veuillez considérer qu'au lieu de $x$, nous pouvons utiliser $X$, qui peut être une représentation polynomiale des $x$. Sous sa forme générale, il serait écrit  
\begin{equation}
y = \log(X)
\end{equation}


```python
X = np.arange(-5.0, 5.0, 0.1)

Y = np.log(X)

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

Une Fonction logistique : 

$$ Y = a + \frac{b}{1+ c^{(X-d)}}$$


```python
X = np.arange(-5.0, 5.0, 0.1)


Y =1-4/(1+np.power(3,X-2))

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

# Exemple de régression non linéaire

# A propos du dataset : 

Ces points de données correspondent au produit intérieur brut (ou PIB) de la Chine de 1960 à 2014.
La première colonne, ce sont les années, et la seconde, le PIB annuel correspondant de la Chine
en dollars américains pour cette année-là.

```python
my_data=pd.read_csv("gdp_of_china.csv")
my_data.head()
```

### Exploration des données 

```python
plt.figure(figsize=(8,5))
x_data, y_data = (my_data["Year"].values, my_data["Value"].values)
plt.plot(x_data, y_data, 'ro')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

Premièrement, "Peut-on prévoir le PIB en fonction du temps ?
Et deuxièmement, "Peut-on utiliser une simple régression linéaire pour le modéliser ?"
En effet, si les données montrent une courbe de tendance, alors la régression linéaire ne produira pas de données très précises
par rapport à une régression non linéaire - simplement parce que, comme son nom l'indique, la régression linéaire
La régression suppose que les données sont linéaires.
Le nuage de points montre qu'il semble y avoir une forte relation entre le PIB et le temps,
mais la relation n'est pas linéaire.
Comme vous pouvez le voir, la croissance démarre lentement, puis à partir de 2005, la croissance est très
significatif.
Et enfin, elle décélère légèrement dans les années 2010.
Cela ressemble un peu à une fonction logistique ou exponentielle.
Elle nécessite donc une méthode d'estimation spéciale de la procédure de régression non linéaire.

### choisir de model 

La figure ci dessus montre que la relation entre le PIB et le temps n'est pas linéaire et qu'il semble à une fonction logistique.

```python
X = np.arange(-5.0, 5.0, 0.1)
Y = 1.0 / (1.0 + np.exp(-X))

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

La formule de la fonction logistique est la suivante :

$$ \hat{Y} = \frac1{1+e^{\beta_1(X-\beta_2)}}$$

1 $ : Contrôle la pente de la courbe,

2 $ : Glisse la courbe sur l'axe des x.

### Construire le model 

```python
def sigmoid(x, Beta_1, Beta_2):
     y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
     return y
```

```python
beta_1 = 0.10
beta_2 = 1990.0

#logistic function
Y_pred = sigmoid(x_data, beta_1 , beta_2)

#prévision initiale du tracé par rapport aux points de données
plt.plot(x_data, Y_pred*15000000000000.)
plt.plot(x_data, y_data, 'ro')
```

Notre tâche consiste ici à trouver les meilleurs paramètres pour notre modèle. Commençons par normaliser nos x et y :

```python
# Normaliser notre données 
xdata =x_data/max(x_data)
ydata =y_data/max(y_data)
```

#### Comment trouver les meilleurs paramètres pour notre ligne d'ajustement ?
nous pouvons utiliser __curve_fit__ qui utilise les moindres carrés non linéaires pour ajuster notre fonction sigmoïde, aux données. Les valeurs optimales des paramètres de sorte que la somme des carrés des résidus de sigmoïde(xdata, *popt) - ydata soit minimisée.

popt sont nos paramètres optimisés.

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, xdata, ydata)
#print the final parameters
print(" beta_1 = %f, beta_2 = %f" % (popt[0], popt[1]))
```

```python
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plt.figure(figsize=(8,5))
y = sigmoid(x, *popt)
plt.plot(xdata, ydata, 'ro', label='data')
plt.plot(x,y, linewidth=3.0, label='fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

### Evaluation 

```python
# split data into train/test
msk = np.random.rand(len(my_data)) < 0.8
train_x = xdata[msk]
test_x = xdata[~msk]
train_y = ydata[msk]
test_y = ydata[~msk]

# build the model using train set
popt, pcov = curve_fit(sigmoid, train_x, train_y)

# predict using test set
y_hat = sigmoid(test_x, *popt)

# evaluation
print("Mean absolute error: %.2f" % np.mean(np.absolute(y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_hat - test_y) ** 2))
from sklearn.metrics import r2_score
print("R2-score: %.2f" % r2_score(y_hat , test_y) )
```
